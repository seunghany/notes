```python
# 기초 그래프 그리기 - Histogram, 산점도 그래프, box graph
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
# Histogram은 column data 분포와 빈도를 살펴보는 용도로 자주 사용하는 그래프 입니다.
# 이때 Dataframe total_bill, tip 등을 변수라고 부르기도 합니다.
# 변수 하나만 사용해서 그린 그래프를 "일변량 그래프" 라고 부릅니다.

# 기본 틀 만들기
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)

# hist method에 total_bil 열을 전달하면 히스토그램이 만들어집니다.
# 이때 x축의 간격은 bins parameter값으로 조정할 수 있습니다.
axes1.hist(tips['total_bill'], bins = 10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

```python
# 이번에는 산점도(Scatter) 그래프를 그려보도록 하겠습니다. 산점도 그래프는 변수 2개를 사용해서 만드는 그래프이며
# 변수 2개를 사용하기 때문에 통계 용어로 '이변량 그래프'라고 부릅니다. 다음은 Total_bill열에 ?따른 tip 열의 분포를
# 나타낸 그래프 입니다.

# 기본 틀
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title("Scatterplot of Total Bill vs Tip")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel('Tip')
```

```python
# 이번에는 박스 그래프를 그려보도록 하겠습니다.
# 박스 그래프는 이산형 변수와 연속형 변수를 함께 사용하는 그래프 입니다.
# 이산형 변수랑 Female, Male과 같이 명확하게 구분되는 값을 의미하고,
# 연속형 변수란 Tip과 같이 명확하게 셀 수 없는 범위의 값을 의미합니다.

# 한번 안보고 해보자

# 먼저 틀 만들기
fig = plt.figure()
axes1 = fig.subplot(1, 1, 1)
axes1.boxplot()
axes1.set_title("Box Plot Graph")
```
